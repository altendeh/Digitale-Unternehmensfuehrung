```python
import os
import pandas as pd
from IPython.display import display

# Verzeichnis, das die extrahierten Dateien enthält
extract_path = 'extracted_files'  # Ersetze dies durch den tatsächlichen Pfad

# Definition der gewünschten Werte in der ersten Spalte
desired_values = {
    "Eigenkapital": ["Eigenkapital"],
    "Umlaufvermögen": ["Summe Aktiva"],
    "Anlagevermögen": ["Immaterielle Vermögensgegenstände", "Sachanlagen", "Finanzanlagen"],
    "Fremdkapital": ["Summe Passiva"]
}

# Dictionary zum Speichern der extrahierten Zeilen
extracted_data = {key: [] for key in desired_values}

# Durchlaufe alle Dateien im Verzeichnis
for file in os.listdir(extract_path):
    if file.endswith('.csv'):
        file_path = os.path.join(extract_path, file)
        try:
            # Einlesen der Datei
            df = pd.read_csv(file_path)
            
            # Filtern der Zeilen basierend auf den gewünschten Werten in der ersten Spalte
            for key, values in desired_values.items():
                filtered_df = df[df.iloc[:, 0].isin(values)]
                
                # Speichern der gefilterten Zeilen
                if not filtered_df.empty:
                    extracted_data[key].append(filtered_df)
                    print(f'Gefilterte Zeilen für {key} in Datei {file}:')
                    display(filtered_df)
        except Exception as e:
            print(f'Fehler beim Einlesen der Datei {file}: {e}')

# Optional: Speichern der extrahierten Daten in separaten CSV-Dateien
for key, data in extracted_data.items():
    if data:
        combined_df = pd.concat(data)
        output_file_path = f'{key.replace(" ", "_").lower()}_data.csv'
        combined_df.to_csv(output_file_path, index=False)
        print(f'Extrahierte Daten für {key} wurden in {output_file_path} gespeichert.')
```

```python
import os
import pandas as pd
from IPython.display import display

# Verzeichnis, das die extrahierten Dateien enthält
extract_path = 'extracted_files'  # Ersetze dies durch den tatsächlichen Pfad

# Definition der gewünschten Werte in der ersten Spalte
desired_values = {
    "Anlagevermögen": ["Immaterielle Vermögensgegenstände", "Sachanlagen", "Finanzanlagen"],
    "Summe Aktiva": ["Summe Aktiva"]
}

# Dictionary zum Speichern der extrahierten Zeilen
extracted_data = {key: [] for key in desired_values}


# Durchlaufe alle Dateien im Verzeichnis
for file in os.listdir(extract_path):
    if file.endswith('.csv'):
        file_path = os.path.join(extract_path, file)
        try:
            # Einlesen der Datei
            df = pd.read_csv(file_path)
            
            # Filtern der Zeilen basierend auf den gewünschten Werten in der ersten Spalte
            for key, values in desired_values.items():
                filtered_df = df[df.iloc[:, 0].isin(values)]
                
                
                # Speichern der gefilterten Zeilen
                if not filtered_df.empty:
                    extracted_data[key].append(filtered_df)
                    print(f'Gefilterte und bereinigte Zeilen für {key} in Datei {file}:')
                    display(filtered_df)
        except Exception as e:
            print(f'Fehler beim Einlesen der Datei {file}: {e}')

# Kombinieren der extrahierten Daten
anlagevermoegen_df = pd.concat(extracted_data["Anlagevermögen"]) if extracted_data["Anlagevermögen"] else pd.DataFrame()
summe_aktiva_df = pd.concat(extracted_data["Summe Aktiva"]) if extracted_data["Summe Aktiva"] else pd.DataFrame()
```

```python
# Funktion zur Bereinigung der Werte in der Spalte "Immaterielle Vermögensgegenstände"
def clean_immaterielle_vermoegensgegenstaende(df):
    if df.iloc[:, 0].str.contains("Immaterielle Vermögensgegenstände").any():
        df.loc[df.iloc[:, 0] == "Immaterielle Vermögensgegenstände", df.columns[1:]] = df.loc[df.iloc[:, 0] == "Immaterielle Vermögensgegenstände", df.columns[1:]].replace({r'\.000$': ''}, regex=True).astype(float)
    return df

# Funktion zur Berechnung der Summe für Anlagevermögen (spaltenweise)
def calculate_anlagevermoegen(df):
    sum_row = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').sum(axis=0)
    sum_row = pd.DataFrame(sum_row).T
    sum_row.insert(0, df.columns[0], 'Summe Anlagevermögen')
    return pd.concat([df, sum_row], ignore_index=True)

# Bereinigung und Berechnung für Anlagevermögen durchführen
if not anlagevermoegen_df.empty:
    anlagevermoegen_df = clean_immaterielle_vermoegensgegenstaende(anlagevermoegen_df)
    anlagevermoegen_df = calculate_anlagevermoegen(anlagevermoegen_df)
    
    # Speichern der berechneten Daten in einer separaten CSV-Datei
    output_file_path = 'anlagevermoegen_data.csv'
    anlagevermoegen_df.to_csv(output_file_path, index=False)
    print(f'Extrahierte und berechnete Daten für Anlagevermögen wurden in {output_file_path} gespeichert.')
    display(anlagevermoegen_df)
else:
    print('Keine Daten für Anlagevermögen gefunden.')
```

```python
# Funktion zur Berechnung des Umlaufvermögens (spaltenweise)
def calculate_umlaufvermoegen(summe_aktiva_df, anlagevermoegen_df):
    summe_aktiva = summe_aktiva_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').sum(axis=0)
    anlagevermoegen = anlagevermoegen_df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').sum(axis=0)
    umlaufvermoegen = summe_aktiva - anlagevermoegen
    umlaufvermoegen_df = pd.DataFrame(umlaufvermoegen).T
    umlaufvermoegen_df.insert(0, summe_aktiva_df.columns[0], 'Umlaufvermögen')
    return umlaufvermoegen_df

# Berechnung für Umlaufvermögen durchführen
if not summe_aktiva_df.empty and not anlagevermoegen_df.empty:
    umlaufvermoegen_df = calculate_umlaufvermoegen(summe_aktiva_df, anlagevermoegen_df)
    
    # Speichern der berechneten Daten in einer separaten CSV-Datei
    output_file_path = 'umlaufvermoegen_data.csv'
    umlaufvermoegen_df.to_csv(output_file_path, index=False)
    print(f'Berechnete Daten für Umlaufvermögen wurden in {output_file_path} gespeichert.')
    display(umlaufvermoegen_df)
else:
    print('Keine ausreichenden Daten für Umlaufvermögen gefunden.')
```

```python
import os
import pandas as pd

# Verzeichnis, das die CSV-Dateien enthält
directory = 'C:/Users/halte/AppData/Local/Temp/b8f3bf0b-88b1-43a5-8b32-8ae28c4fafa9_output.zip.fa9'

# Definition der gewünschten Zeilen
desired_rows = {
    "Eigenkapital": ["Eigenkapital"],
    "Umlaufvermögen": ["Summe Aktiva"],
    "Anlagevermögen": ["Immaterielle Vermögensgegenstände", "Sachanlagen", "Finanzanlagen"],
    "Fremdkapial": ["Summe Passiva"]
}

# Dictionary zum Speichern der extrahierten Zeilen
extracted_data = {key: [] for key in desired_rows}
headers = {key: None for key in desired_rows}

# Durchlaufe alle Dateien im Verzeichnis
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        file_path = os.path.join(directory, filename)
        
        # Datei zeilenweise einlesen
        with open(file_path, 'r', encoding='utf-8') as file:
            lines = file.readlines()
        
        # Extrahieren der Header-Zeile
        header = lines[0].strip()
        
        # Extrahieren der gewünschten Zeilen
        for line in lines[1:]:  # Überspringe die Header-Zeile
            for key, row_names in desired_rows.items():
                if any(row_name.lower() in line.lower() for row_name in row_names):
                    extracted_data[key].append(line.strip())
                    headers[key] = header
                    print(f'Gefundene Zeile für {key} in Datei {filename}: {line.strip()}')

# Summieren der Werte für die gewünschten Zeilen
summed_data = {key: [] for key in desired_rows}
for key, data in extracted_data.items():
    if data:
        # Konvertiere die Zeilen in DataFrame
        df = pd.DataFrame([x.split(',') for x in data])
        # Summiere die numerischen Spalten
        summed_row = df.iloc[:, 1:].apply(pd.to_numeric, errors='coerce').sum().fillna(0)
        # Füge den Zeilennamen hinzu
        summed_row = pd.Series([key] + summed_row.tolist())
        summed_data[key].append(summed_row)

# Speichern der extrahierten Zeilen und der summierten Daten in separaten CSV-Dateien
for key, data in extracted_data.items():
    output_file_path = f'{key.replace(" ", "_").lower()}_data.csv'
    with open(output_file_path, 'w', encoding='utf-8') as file:
        # Schreibe den Header
        if headers[key]:
            file.write(headers[key] + '\n')
        # Schreibe die extrahierten Zeilen
        for line in data:
            file.write(line + '\n')
    print(f'Extrahierte Daten für {key} wurden in {output_file_path} gespeichert.')

# Speichern der summierten Daten in separaten CSV-Dateien
for key, data in summed_data.items():
    output_file_path = f'{key.replace(" ", "_").lower()}_summed_data.csv'
    with open(output_file_path, 'w', encoding='utf-8') as file:
        # Schreibe den Header
        if headers[key]:
            file.write(headers[key] + '\n')
        # Schreibe die summierten Zeilen
        for line in data:
            file.write(','.join(map(str, line)) + '\n')
    print(f'Summierte Daten für {key} wurden in {output_file_path} gespeichert.')
```
